Importing the libraries

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
```

Importing the dataset

```python
#loading the dataset to a Pandas DataFrame
credit_card_data=pd.read_csv('creditcard.csv')
```

```python
#first 5 rows of the dataset
credit_card_data.head()
```

```python
credit_card_data.tail()
```

```python
#dataset informations
credit_card_data.info()
```

```python
#checking the number of missing values in each column
credit_card_data.isnull().sum()
```

```python
credit_card_data['V3'].fillna(credit_card_data['V3'].median(),inplace=True)
```

```python
credit_card_data.tail()
```

```python
#removing the null values using imputation

num_columns = len(credit_card_data.columns)

for i in range(5, num_columns):
    credit_card_data.iloc[:, i].fillna(credit_card_data.iloc[:, i].median(), inplace=True)
```

```python
credit_card_data.isnull().sum()
```

```python
credit_card_data.tail()
```

```python
#distribution of legit trnasaction and fraudulent transactions
credit_card_data['Class'].value_counts()
```

This dataset is highly unbalanced

0 -> Normal Transactions
1 -> Fraudulent Transactions

```python
#separating the data for analysis
legit = credit_card_data[credit_card_data.Class==0]
fraud = credit_card_data[credit_card_data.Class==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#statistical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#compare the values for both transactions
credit_card_data.groupby('Class').mean()
```

Under-Sampling

Build a sample dataset containing similar distribution of Normal transactions and Fraudulent transactions

Number of Fraudulent Transactions -> 38

```python
legit_sample=legit.sample(n=472)
```

Concatenating two DataFrames

```python
new_dataset=pd.concat([legit_sample,fraud],axis=0)
#axis=0 means adding values row wise
#and axis=1 means adding values column wise
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
#compare the values for both transactions
new_dataset.groupby('Class').mean()
```

Splitting the data into Features & Targets

```python
X=new_dataset.drop(columns='Class',axis=1)
Y=new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

Split the data into Training Data and 
Testing Data

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
#stratify is used for the even distribution
#random state is used to produce the data as it is
```

```python
print(X_train)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
print(X_test)
```

```python
print(Y_train)
```

```python
print(Y_test)
```

Model Training

Logistic Regrssion(because of binary classification problem)

```python
model=LogisticRegression()
```

```python
#training the Logistic Regression Model with Training Data
model.fit(X_train,Y_train)
```

Model Evaluation

Accuracy Score

```python
#accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
#predicting all the labels of training data and stores them into X_train_prediction
#and then compares it with labels of Y_train(original labels)
```

```python
print('Accuracy on Trainig data : ', training_data_accuracy)
```

```python
#accuracy on test data
X_test_prediction=model.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction,Y_test)
```

```python
print('Accuracy on Test data : ', test_data_accuracy)
```
